# Import Libraries

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import numpy as np
```

```python
def random_data_visualize(ad_path, other_path):
    ad_path = (ad_path)
    num_ad = len(os.listdir(ad_path))
    print('Atopic Dermatitis', num_ad)
    
    print('')
    
    other_path = (other_path)
    num_other = len(os.listdir(other_path))
    print('Atopic Dermatitis', num_ad)
    
    classes = ['Atopic Dermatits', 'Other']
    number_image = [num_ad, num_other]
    num_classes = np.arange(len(classes))
    plt.figure(figsize=(12, 4))
    plt.bar(num_classes, number_image)
    plt.xticks(num_classes, classes, fontsize=15)
    plt.xlabel('Classes')
    plt.ylabel('Number of Image')
    plt.show()
```
