# Visualize

Visualize is important

```python
import matplotlib.pyplot as plt
import glob
import matplotlib.image as mpimg
import os
import numpy as np
```

# Arm

```python
ad_arm_path = './Data/data_step1_crop/AtopicDermatitis/arm/*.jpg'
cd_arm_path = './Data/data_step1_crop/ContactDermatitis/arm/*.jpg'
```

```python
ad_arm_images = []
for images in glob.glob(ad_arm_path):
    ad_arm_images.append(mpimg.imread(images))

plt.figure(figsize=(30,20))
columns = 5
for i, image in enumerate(ad_arm_images):
    plt.subplot(len(images)/columns + 1, columns, i + 1)
    plt.imshow(image)
plt.title('Atopic Dermatitis')
```

```python
cd_arm_images = []
for images in glob.glob(cd_arm_path):
    cd_arm_images.append(mpimg.imread(images))

plt.figure(figsize=(30,20))
columns = 5
for i, image in enumerate(cd_arm_images):
    plt.subplot(len(images)/columns + 1, columns, i + 1)
    plt.imshow(image)
plt.title('Contact Dermatitis')
```

```python
print('Number of Arm Images')

ad_arm_path = './Data/data_step1_crop/AtopicDermatitis/arm/'
num_ad_arm = len(os.listdir(ad_arm_path))-1
print('Atopic Dermatitis', num_ad_arm)

cd_arm_path = './Data/data_step1_crop/ContactDermatitis/arm/'
num_cd_arm = len(os.listdir(cd_arm_path))-1
print('Contact Dermatitis', num_cd_arm)

classes = ['Atopic Dermatits', 'Contact Dermatitis']
number_image = [num_ad_arm, num_cd_arm]
num_classes = np.arange(len(classes))

plt.bar(num_classes, number_image)
plt.xticks(num_classes, classes, fontsize=15)
plt.xlabel('Classes')
plt.ylabel('Number of Picture')
plt.show()
```
